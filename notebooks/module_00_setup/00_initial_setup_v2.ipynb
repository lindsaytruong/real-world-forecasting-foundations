# Initial Setup

In this lesson, we will focus on ensuring our environment is ready to go before we start jumping in to any data.
In this lesson we will.
1. Install VS Code - the editor we will use throughout the course
2. Install a clean Python environment for the course, so your system stays tidy.  
3. Add the VS Code juypyter + python extensions
4. Create the course environment from an environment.yml file
5. Make the environment show up as a jupyter kernel in VS COde 
6.  Run a quick "Hello Nixtla" check to confirm it works.

By the end, you'll be able to open a notebook in VS code and run Nixtla code in one click.

## 1. Install VS Code

The first thing we need to do is ensure we have VSCOde installed.  If you do not, 
1. Go to https://code.visualstudio.com/ and click download VS code for your OS. 
2. Install with the defaults
3. Launch VS Code once it finishes first-run setup.

If you’re using Conda or Mamba (which we’ll install in the next step), you do not need to install standalone Python from python.org. We’ll create our own Python environment inside Conda that VS Code will use automatically.

### Next we have to install a couple extensions in VSCode.
1. Go over the extensions on the side panel and search for 
- Python  by microsoft - and then click install
- Jupyter by Microsoft - and click install

## 2. Install Conda

If you already have Anaconda, Miniconda, or Mambaforge installed, you can skip the installation step. Just open a terminal and type
```conda --version```
If that prints a version number, you’re good to go. Move ahead to Step 3: Get the environment file.”
That ensures you’re not forcing unnecessary re-installs.
Whether you use Anaconda or Miniconda doesn’t matter — both can create the same environment file. If you use full Anaconda, your base environment may be large, but our course env (fa-nixtla) will stay lightweight and isolated.

I will show how to install miniconda

Download Miniconda from https://docs.conda.io/en/latest/miniconda.html
 (pick your OS).

Install with defaults. Allow “Add to PATH” if prompted on Windows.

- Close/reopen your terminal after installing Miniconda/Mambaforge. On Windows, use “Anaconda Prompt”.

## 3. Clone the repo

You will need the environment.yml that defines our course environment.

You can either download this from our course github, or from course page.

I will demonstrate the github path, and to do that, you will need to clone the repo.
So 
1. open the terminal and type ```git clone https://github.com/<your-org>/<course-repo>.git```
2. then ```cd <course repo>```
3. if you type ```ls``` you should see the environment.yml in your files



## 4. Install the environment.yml

- Open a terminal:
- Windows (PowerShell): open “Anaconda Prompt” or “PowerShell”
- macOS/Linux: open Terminal
- Navigate to the folder containing environment.yml, then run:

```conda env create -f environment.yml```
 - or, if you use mamba: ```mamba env create -f environment.yml```

- activate it
``` conda activate fa-nixtla```
Register it as a Jupyter kernel (so VS Code can see it):
```python -m ipykernel install --user --name fa-nixtla --display-name "Python (fa-nixtla)```

After installing ipykernel, you will need to restart VSCode.

Updating later?
If I ship a new environment.yml, you can update with:
```conda env update -f environment.yml --prune```

## 5. Connect VS Code to your environment

Open the course folder in VS Code (File → Open Folder…).

- Select the interpreter:
- Press Ctrl/Cmd + Shift + P → type “Python: Select Interpreter”.
- Choose the one that shows fa-nixtla (Conda).
- Open a notebook (.ipynb) or create one (New File → Jupyter Notebook).
- Pick the kernel: top-right of the notebook, click the kernel name and choose Python (fa-nixtla).

You’re now running notebooks with the course environment.
